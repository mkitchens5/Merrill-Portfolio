```python
from pandas import Series, DataFrame

movies_data['genres'].str.split('|').apply(Series,1).stack()
```

```python
#assigning each of the genre's in every row to its appropiate index and orginal_title
split = movies_data['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_data['genres']
movies_data = movies_data.join(split)
```

### Research Questions (Q):
2. Which genre's have high avg. budget and revenue?

```python
# Writing code to visualize the data from all the averages per genre
genres_avg = movies_data.groupby('genres_split').mean(numeric_only=True)

#Changing visualization numbers to return as a float instead of the numbers being in scientific notation 
pd.options.display.float_format = '{:2f}'.format


#This graph shows the visualization of data of all the averages per genre
genres_avg
```

```python
#Sorting values
genres_avg.sort_values('budget', ascending = True, inplace = True)
```

```python
# 2. Which genre's have high avg. budget and revenue? (This code answers the question for highest avg budget to lowest in descending order)
genres_avg[['budget', 'revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('DarkBlue','c'), figsize = (10,9))
```

```python
#Sorting values
genres_avg.sort_values('revenue', ascending = True, inplace = True)

# 2. Which genre's have high avg. budget and revenue? (This code answers the question for highest avg revenue to lowest in descending order)
genres_avg[['budget', 'revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('DarkBlue','c'), figsize = (10,9))
```

### Research Questions (Q):
